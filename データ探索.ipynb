```python
import pandas as pd
url="http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"
df=pd.read_excel(url)
df
```

```python
import numpy as np
df_2=df[np.isfinite(df['CustomerID'])]
df_2
```

```python
df_3=df_2[~df_2['InvoiceNo'].str.contains("C",na=False)]
df_3
```

```python
a=df_3['CustomerID'].drop_duplicates()
a
```

```python
b=df_3['InvoiceNo'].drop_duplicates()
b
```

```python
c=df_3['StockCode'].drop_duplicates()
c
```

```python
d=df_3['Description'].drop_duplicates()
d
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
df_3["z"]=df_3["Quantity"]*df_3["UnitPrice"]
df_3["z"]
```
